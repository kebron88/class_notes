```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dates = [1900,1910,1920,1930,1940,1950,1960,1970,1980,1990,2000]
males_heights = [162,163,165,167,168,169,171,172,172,172,171]
females_heights = [151,153,154,156,158,159,159,160,160,160,159]
```

```python
plt.plot(dates, males_heights)
plt.show()
```

```python
#styling
plt.style.use('ggplot')
#multiple plots on the same figure
plt.plot(dates, males_heights, label='Male')#, color=(0.3,0.9,0.6), linewidth=6)
plt.plot(dates, females_heights, label='Female')
#legend
plt.legend()
# plt.gca().set_facecolor('lightgrey')
# plt.grid()
plt.title('Height vs Year')
plt.xlabel('Year')
plt.ylabel('Avg Height')
plt.show()
```

```python
fig, axes = plt.subplots(1,2,figsize=(8,4))
axes[0].plot(dates, males_heights)
axes[1].plot(dates, females_heights)
axes[0].set_title('Males Heights vs Year')
axes[0].set_xlabel('Date')
plt.show()
```

```python
df = pd.read_csv('iris.csv')
df.sample(10)
```

```python
plt.style.use('default')
plt.scatter(df['petal_length'], df['sepal_width'], color='green', s=df.petal_width*10, alpha=0.5)
plt.title('sepal width vs petal length')
plt.xlabel('petal length')
plt.ylabel('sepal width')
plt.show()
```

```python
df1 = df[df.species == 'Iris-setosa']
df2 = df[df.species == 'Iris-versicolor']
df3 = df[df.species == 'Iris-virginica']
plt.scatter(df1.petal_length, df1.sepal_width, color='red', label='setosa')
plt.scatter(df2.petal_length, df2.sepal_width, color='green', label='versicolor')
plt.scatter(df3.petal_length, df3.sepal_width, color='blue', label='virginica')
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(1,3, figsize=(10,3))
axes[0].scatter(df1.petal_length, df1.sepal_width, color='red', label='setosa')
axes[1].scatter(df2.petal_length, df2.sepal_width, color='green', label='versicolor')
axes[2].scatter(df3.petal_length, df3.sepal_width, color='blue', label='virginica')
plt.show()
```

```python
plt.hist(df1.sepal_length, bins=10, color='red', alpha=0.5, label='setosa')
plt.hist(df2.sepal_length, bins=10, color='green', alpha=0.5, label='versicolor')
plt.hist(df3.sepal_length, bins=10, color='blue', alpha=0.5, label='virginica')
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(1,3, figsize=(10,3))
axes[0].hist(df1.sepal_length, bins=10, color='red', alpha=0.5, label='setosa')
axes[1].hist(df2.sepal_length, bins=10, color='green', alpha=0.5, label='versicolor')
axes[2].hist(df3.sepal_length, bins=10, color='blue', alpha=0.5, label='virginica')
plt.show()
```
