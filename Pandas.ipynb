```python
df1 = pd.DataFrame([[1,'Alice'],[2,'Bob'],[3,'Cathy']], columns=['ID','Name'])
df2 = pd.DataFrame([[1,'Washington'],[2,'Oregon'],[4,'California']], columns=['ID','State'])
df3 = pd.DataFrame([[4,'Dave'],[5,'Emma'],[6,'Frank']], columns=['ID','Name'])
print('df1')
print(df1)
print()
print('df2')
print(df2)
print()
print('df3')
print(df3)
print()
```

```python
#join df2 onto df1 using the 'ID' column as the common column between them. Keep all the rows where the ID is 
#in both tables as well as the rows where the ID is only in df2
df1.merge(df2, on='ID', how='right')
```

```python
#combine the rows of df1 and df3
pd.concat([df1,df3])
```

```python
#subset the rows by a custom condition. If the function evaluates to true when applied on the column, the row is kept
#and dropped if it evaluates to False. 
def short_name(s):
    if len(s) <= 15:
        return True
    else:
        return False 
    
df[df['Full Name'].apply(short_name)]
```
